```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
titanic_data = pd.read_csv("/content/titanicdataset.csv")
```

```python
# Display the first few rows of the dataset
print(titanic_data.head())
```

```python
# Display summary statistics for numerical columns
print(titanic_data.describe())
```

```python
# Display information about the dataset, including data types and missing values
print(titanic_data.info())
```

```python
# Check for missing values in each column
print(titanic_data.isnull().sum())
```

```python
# Drop rows with missing values in the "Embarked" column
titanic_data.dropna(subset=["Embarked"], inplace=True)
```

```python
# Fill missing values in the "Cabin" column with "Unknown"
titanic_data["Cabin"].fillna("Unknown", inplace=True)
```

```python
# Fill missing values in the "Age" column with the mean age
titanic_data["Age"].fillna(titanic_data["Age"].mean(), inplace=True)
```

```python
# Check for missing values again after cleaning
print(titanic_data.isnull().sum())
```

```python
# Check for duplicated rows
print(titanic_data.duplicated().sum())
```

```python
# Visualization: Box Plot of Age
plt.figure(figsize=(8, 6))
sns.boxplot(x=titanic_data["Age"], color='darksalmon')
plt.title("Distribution of Passenger Ages", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.show()
```

```python
# Visualization: Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(titanic_data["Age"], kde=True, color='darksalmon')
plt.title("Distribution of Passenger Ages", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80], fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Visualization: Survival by Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_data, x="Sex", hue="Survived", palette='coolwarm')
plt.title("Survival by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Survived", loc="upper right")
plt.yticks([0, 50, 100, 150, 200, 250, 300, 350, 400], fontsize=10)
plt.xticks(fontsize=10)
plt.show()
```

```python


# Visualization: Scatter Plot of Age vs Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic_data, x="Age", y="Fare", hue="Survived", palette='coolwarm')
plt.title("Scatter Plot of Age vs Fare", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.legend(title="Survived")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
